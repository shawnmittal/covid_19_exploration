```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
directory = Path("../COVID-19/csse_covid_19_data/csse_covid_19_time_series")
conf_file = Path("time_series_19-covid-Confirmed.csv")
death_file = Path("time_series_19-covid-Deaths.csv")
recov_file = Path("time_series_19-covid-Recovered.csv")
```

```python
confirmed = pd.read_csv(directory / conf_file)
deaths = pd.read_csv(directory / death_file)
recov = pd.read_csv(directory / recov_file)
```

```python
plt.figure(figsize=[20,10])
plt.plot(confirmed[confirmed["Country/Region"] != "Mainland China"].sum()[3:])
plt.plot(deaths[deaths["Country/Region"] != "Mainland China"].sum()[3:])
plt.plot(recov[recov["Country/Region"] != "Mainland China"].sum()[3:])
plt.legend(["confirmed", "deaths", "recovered"])
plt.title("COVID-19 Statistics Outside of Mainland China")
plt.xlabel("Date")
plt.ylabel("Population Count")
plt.xticks(rotation=45);
```

```python
plt.figure(figsize=[20,10])
plt.plot(confirmed[confirmed["Country/Region"] == "Mainland China"].sum()[4:])
plt.plot(deaths[deaths["Country/Region"] == "Mainland China"].sum()[4:])
plt.plot(recov[recov["Country/Region"] == "Mainland China"].sum()[4:])
plt.legend(["confirmed", "deaths", "recovered"])
plt.title("COVID-19 Statistics Inside of Mainland China")
plt.xlabel("Date")
plt.ylabel("Population Count")
plt.xticks(rotation=45);
```

```python
plt.figure(figsize=[20,10])
plt.plot(confirmed[confirmed["Country/Region"] == "Mainland China"].sum()[4:])
plt.plot(deaths[deaths["Country/Region"] == "Mainland China"].sum()[4:])
plt.plot(recov[recov["Country/Region"] == "Mainland China"].sum()[4:])
plt.plot(confirmed[confirmed["Country/Region"] != "Mainland China"].sum()[3:])
plt.plot(deaths[deaths["Country/Region"] != "Mainland China"].sum()[3:])
plt.plot(recov[recov["Country/Region"] != "Mainland China"].sum()[3:])
plt.legend(["confirmed-MC", "deaths-MC", "recovered-MC", "confirmed-OMC", "deaths-OMC", "recovered-OMC"])
plt.title("COVID-19 Statistics Inside of Mainland China")
plt.xlabel("Date")
plt.ylabel("Population Count")
plt.xticks(rotation=45);
```
